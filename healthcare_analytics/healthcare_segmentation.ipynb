```python
# Healthcare Patient Segmentation Analysis
# Author: [Your Name]
# Date: March 13, 2025

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Import the RFM analysis module using standard import
# This assumes the file has been renamed to rfm_analysis.py
from rfm_analysis import HealthcareRFM

# Set visualization styles
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)

# 1. Data Loading
print("Loading healthcare data...")

# Sample healthcare data - replace with your actual data path
try:
    df = pd.read_csv('../data/healthcare_data.csv')
    print("Data loaded successfully!")
except FileNotFoundError:
    # Generate synthetic data for demonstration
    print("Data file not found. Generating synthetic data for demonstration...")
    
    np.random.seed(42)
    n_samples = 1000
    
    # Generate synthetic patient data
    patient_ids = [f"P{i:04d}" for i in range(1, n_samples + 1)]
    age = np.random.normal(55, 15, n_samples).astype(int)
    age = np.clip(age, 18, 90)
    
    # Generate gender (0: Male, 1: Female)
    gender = np.random.binomial(1, 0.55, n_samples)
    
    # Generate BMI with realistic distribution
    bmi = np.random.normal(28, 5, n_samples)
    bmi = np.clip(bmi, 16, 45)
    
    # Generate blood pressure related features
    systolic_bp = np.random.normal(125, 15, n_samples)
    diastolic_bp = np.random.normal(80, 10, n_samples)
    
    # Generate blood test results
    blood_glucose = np.random.normal(100, 25, n_samples)
    cholesterol = np.random.normal(190, 35, n_samples)
    hdl = np.random.normal(50, 15, n_samples)
    ldl = np.random.normal(120, 30, n_samples)
    
    # Generate healthcare utilization metrics
    annual_med_costs = np.random.exponential(5000, n_samples)
    hospital_visits = np.random.poisson(2, n_samples)
    emergency_visits = np.random.poisson(0.8, n_samples)
    prescription_count = np.random.poisson(5, n_samples)
    
    # Generate chronic conditions counts (0-5)
    chronic_conditions = np.random.binomial(5, 0.3, n_samples)
    
    # Generate visit dates for RFM analysis
    visit_dates = []
    visit_costs = []
    
    for i in range(n_samples):
        num_visits = max(1, np.random.poisson(hospital_visits[i] + 2))
        
        # Base date with some randomness
        if age[i] > 65:
            # Older patients may have more recent visits
            days_ago = np.random.randint(0, 365)
        else:
            days_ago = np.random.randint(0, 730)
            
        latest_visit = datetime.now() - pd.Timedelta(days=days_ago)
        patient_visit_dates = [latest_visit]
        
        # Generate earlier visits with decreasing probability
        for j in range(num_visits - 1):
            days_gap = np.random.randint(30, 180)
            visit_date = patient_visit_dates[-1] - pd.Timedelta(days=days_gap)
            patient_visit_dates.append(visit_date)
        
        visit_dates.extend(patient_visit_dates)
        
        # Generate costs that correlate with chronic conditions
        base_cost = 500 + chronic_conditions[i] * 300
        visit_costs.extend([base_cost + np.random.gamma(3, 200) for _ in range(num_visits)])
    
    # Create the DataFrame
    data = {
        'patient_id': patient_ids,
        'age': age,
        'gender': gender,
        'bmi': bmi,
        'systolic_bp': systolic_bp,
        'diastolic_bp': diastolic_bp,
        'blood_glucose': blood_glucose,
        'cholesterol': cholesterol,
        'hdl': hdl,
        'ldl': ldl,
        'annual_med_costs': annual_med_costs,
        'hospital_visits': hospital_visits,
        'emergency_visits': emergency_visits,
        'prescription_count': prescription_count,
        'chronic_conditions': chronic_conditions
    }
    
    df = pd.DataFrame(data)
    
    # Create a separate DataFrame for visit-level data for RFM analysis
    visits_data = []
    for i, pid in enumerate(patient_ids):
        num_visits = max(1, np.random.poisson(hospital_visits[i] + 2))
        for _ in range(num_visits):
            visit_date = datetime.now() - pd.Timedelta(days=np.random.randint(0, 730))
            visit_cost = 200 + chronic_conditions[i] * 100 + np.random.gamma(3, 100)
            visits_data.append({
                'patient_id': pid,
                'visit_date': visit_date,
                'visit_cost': visit_cost
            })
    
    visit_df = pd.DataFrame(visits_data)
    visit_df = visit_df.sort_values(['patient_id', 'visit_date'])
    
    print("Synthetic data generated successfully!")

# 2. Exploratory Data Analysis
print("\n--- Exploratory Data Analysis ---")

# Display basic information about the dataset
print("\nDataset Overview:")
print(f"Shape: {df.shape}")
print(f"Number of patients: {df.shape[0]}")
print(f"Number of features: {df.shape[1]}")

# Display the first few rows of the dataset
print("\nFirst 5 rows of the dataset:")
display(df.head())

# Display summary statistics
print("\nSummary Statistics:")
display(df.describe())

# Check for missing values
print("\nMissing Values:")
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percent})
display(missing_data[missing_data['Missing Values'] > 0])

# 3. Data Preprocessing
print("\n--- Data Preprocessing ---")

# Handle missing values (if any)
if df.isnull().sum().sum() > 0:
    print("Handling missing values...")
    # For numerical columns: fill with median
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in num_cols:
        if df[col].isnull().sum() > 0:
            df[col].fillna(df[col].median(), inplace=True)
    
    # For categorical columns: fill with mode
    cat_cols = df.select_dtypes(include=['object', 'category']).columns
    for col in cat_cols:
        if df[col].isnull().sum() > 0:
            df[col].fillna(df[col].mode()[0], inplace=True)
    
    print("Missing values handled.")

# Convert gender to numeric if it's not already
if df['gender'].dtype == 'object':
    df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
    print("Gender converted to numeric.")

# Create age groups for analysis
age_bins = [18, 35, 50, 65, 80, 100]
age_labels = ['18-34', '35-49', '50-64', '65-79', '80+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
print("Age groups created.")

# Create BMI categories
bmi_bins = [0, 18.5, 25, 30, 35, 100]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese I', 'Obese II+']
df['bmi_category'] = pd.cut(df['bmi'], bins=bmi_bins, labels=bmi_labels, right=False)
print("BMI categories created.")

# 4. Feature Analysis
print("\n--- Feature Analysis ---")

# Age distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.axvline(df['age'].mean(), color='red', linestyle='--', label=f'Mean: {df["age"].mean():.1f}')
plt.legend()
plt.show()

# Gender distribution
plt.figure(figsize=(8, 6))
gender_counts = df['gender'].value_counts()
gender_labels = ['Male', 'Female'] if sorted(df['gender'].unique()) == [0, 1] else df['gender'].unique()
plt.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution', fontsize=16)
plt.axis('equal')
plt.show()

# BMI distribution by age group
plt.figure(figsize=(12, 8))
sns.boxplot(x='age_group', y='bmi', data=df)
plt.title('BMI Distribution by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('BMI', fontsize=12)
plt.axhline(y=25, color='red', linestyle='--', label='Overweight Threshold')
plt.axhline(y=30, color='darkred', linestyle='--', label='Obesity Threshold')
plt.legend()
plt.show()

# Medical costs analysis
plt.figure(figsize=(12, 8))
sns.scatterplot(x='age', y='annual_med_costs', hue='chronic_conditions', size='hospital_visits', 
                sizes=(20, 200), alpha=0.7, palette='viridis', data=df)
plt.title('Annual Medical Costs by Age and Chronic Conditions', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Annual Medical Costs ($)', fontsize=12)
plt.legend(title='Chronic Conditions Count')
plt.show()

# Correlation heatmap
plt.figure(figsize=(14, 10))
# Select only numeric columns for correlation
numeric_cols = ['age', 'gender', 'bmi', 'systolic_bp', 'diastolic_bp', 'blood_glucose', 
                'cholesterol', 'hdl', 'ldl', 'annual_med_costs', 'hospital_visits', 
                'emergency_visits', 'prescription_count', 'chronic_conditions']
numeric_df = df[numeric_cols]
corr = numeric_df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5, 
            vmin=-1, vmax=1, center=0, square=True)
plt.title('Correlation Heatmap of Healthcare Features', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. K-Means Clustering for Patient Segmentation
print("\n--- K-Means Clustering for Patient Segmentation ---")

# Prepare data for clustering
# Select relevant features for clustering
clustering_features = ['age', 'bmi', 'systolic_bp', 'diastolic_bp', 'blood_glucose', 
                       'cholesterol', 'annual_med_costs', 'hospital_visits', 
                       'emergency_visits', 'prescription_count', 'chronic_conditions']

X = df[clustering_features].copy()

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Data standardized for clustering.")

# Apply PCA for dimensionality reduction and visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print(f"PCA explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance explained: {sum(pca.explained_variance_ratio_):.2f}")

# Determine optimal number of clusters using the elbow method and silhouette score
print("\nDetermining optimal number of clusters...")
wcss = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X_scaled, labels))

# Plot the elbow method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method for Optimal k', fontsize=14)
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('WCSS', fontsize=12)
plt.grid(True)

# Plot silhouette scores
plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score Method for Optimal k', fontsize=14)
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('Silhouette Score', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Find the best number of clusters based on silhouette score
best_k = k_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Optimal number of clusters based on silhouette score: {best_k}")

# Apply K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['kmeans_cluster'] = kmeans.fit_predict(X_scaled)
print(f"K-means clustering applied with {best_k} clusters.")

# Visualize the clusters in PCA space
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['kmeans_cluster'], cmap='viridis', alpha=0.8, s=50)
plt.title(f'Patient Segments Visualization Using PCA (k={best_k})', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.show()

# 6. DBSCAN Clustering for Identifying Unusual Patient Groups
print("\n--- DBSCAN Clustering for Identifying Unusual Patient Groups ---")

# Apply DBSCAN clustering
# DBSCAN parameters need to be tuned based on the data
eps = 0.5  # Maximum distance between two samples to be considered in the same neighborhood
min_samples = 5  # Minimum number of samples in a neighborhood to form a core point

dbscan = DBSCAN(eps=eps, min_samples=min_samples)
df['dbscan_cluster'] = dbscan.fit_predict(X_scaled)

# Count clusters (including noise points which are labeled as -1)
n_clusters = len(set(df['dbscan_cluster'])) - (1 if -1 in df['dbscan_cluster'] else 0)
n_noise = list(df['dbscan_cluster']).count(-1)

print(f"Number of clusters found by DBSCAN: {n_clusters}")
print(f"Number of noise points: {n_noise} ({n_noise/len(df)*100:.1f}%)")

# Visualize DBSCAN clusters in PCA space
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['dbscan_cluster'], cmap='viridis', alpha=0.8, s=50)
plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples})', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 7. Hierarchical Clustering
print("\n--- Hierarchical Clustering ---")

# Perform hierarchical clustering
# Use a sample for dendrogram visualization if dataset is large
if len(df) > 200:
    # Take a random sample for visualization
    sample_idx = np.random.choice(len(df), 200, replace=False)
    sample_data = X_scaled[sample_idx]
    sample_labels = [f"P{i}" for i in range(len(sample_data))]
else:
    sample_data = X_scaled
    sample_labels = [f"P{i}" for i in range(len(X_scaled))]

# Compute the linkage matrix
Z = linkage(sample_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(15, 8))
plt.title('Hierarchical Clustering Dendrogram', fontsize=16)
plt.xlabel('Patient Index', fontsize=12)
plt.ylabel('Distance', fontsize=12)
dendrogram(
    Z,
    truncate_mode='lastp',  # Show only the last p clusters
    p=10,  # Show only the last 10 merges
    leaf_rotation=90.,
    leaf_font_size=10.,
    labels=sample_labels
)
plt.tight_layout()
plt.show()

# 8. RFM Analysis Integration
print("\n--- RFM Analysis Integration ---")

# Check if we have visit_df (from synthetic data) or need to create it
try:
    print(f"Visit data shape: {visit_df.shape}")
except NameError:
    # Create sample visit data if it doesn't exist
    print("Generating visit data for RFM analysis...")
    visits_data = []
    for i, pid in enumerate(df['patient_id']):
        num_visits = max(1, df.loc[i, 'hospital_visits'])
        for _ in range(int(num_visits)):
            visit_date = datetime.now() - pd.Timedelta(days=np.random.randint(0, 730))
            visit_cost = df.loc[i, 'annual_med_costs'] / num_visits
            visits_data.append({
                'patient_id': pid,
                'visit_date': visit_date,
                'visit_cost': visit_cost
            })
    
    visit_df = pd.DataFrame(visits_data)
    visit_df = visit_df.sort_values(['patient_id', 'visit_date'])
    print(f"Generated {len(visit_df)} visits for RFM analysis")

# Initialize RFM analyzer
rfm_analyzer = HealthcareRFM()

# Perform RFM analysis
print("Performing RFM analysis...")
rfm_result = rfm_analyzer.analyze(
    data=visit_df,
    patient_id_col='patient_id',
    date_col='visit_date',
    monetary_col='visit_cost',
    reference_date=datetime.now()
)

# Segment patients based on RFM scores
segmented_rfm = rfm_analyzer.segment_patients(rfm_result)

# Display RFM segment distribution
print("\nRFM Segment Distribution:")
segment_distribution = rfm_analyzer.get_segment_distribution(segmented_rfm)
print(segment_distribution)

# Plot RFM segment distribution
rfm_analyzer.plot_segment_distribution(segmented_rfm)

# Plot RFM heatmap
rfm_analyzer.plot_rfm_heatmap(segmented_rfm)

# Merge RFM segments with cluster analysis
merged_results = df.merge(
    segmented_rfm[['patient_id', 'segment', 'recency_score', 'frequency_score', 'monetary_score']],
    on='patient_id',
    how='left'
)

# Rename the segment column to avoid confusion
merged_results.rename(columns={'segment': 'rfm_segment'}, inplace=True)

# 9. Analyze the relationship between K-means clusters and RFM segments
print("\n--- K-means Clusters vs. RFM Segments ---")

# Create a cross-tabulation
cross_tab = pd.crosstab(
    merged_results['kmeans_cluster'], 
    merged_results['rfm_segment'],
    normalize='index'
) * 100  # Convert to percentages

print("\nPercentage of RFM segments within each K-means cluster:")
display(cross_tab)

# Visualize the relationship
plt.figure(figsize=(14, 10))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Distribution of RFM Segments within K-means Clusters (%)', fontsize=14)
plt.xlabel('RFM Segment', fontsize=12)
plt.ylabel('K-means Cluster', fontsize=12)
plt.tight_layout()
plt.show()

# 10. Cluster Interpretation and Profiling
print("\n--- Cluster Interpretation and Profiling ---")

# Profile K-means clusters
kmeans_profiles = merged_results.groupby('kmeans_cluster').agg({
    'age': 'mean',
    'bmi': 'mean',
    'systolic_bp': 'mean',
    'diastolic_bp': 'mean',
    'blood_glucose': 'mean',
    'cholesterol': 'mean',
    'annual_med_costs': 'mean',
    'hospital_visits': 'mean',
    'emergency_visits': 'mean',
    'chronic_conditions': 'mean',
    'recency_score': 'mean',
    'frequency_score': 'mean',
    'monetary_score': 'mean',
    'patient_id': 'count'
}).rename(columns={'patient_id': 'count'})

print("\nK-means Cluster Profiles:")
display(kmeans_profiles)

# Visualize key metrics across clusters
plt.figure(figsize=(15, 12))
metrics = ['age', 'bmi', 'annual_med_costs', 'hospital_visits', 'chronic_conditions', 'recency_score', 'frequency_score', 'monetary_score']
n_rows = (len(metrics) + 1) // 2  # Calculate number of rows needed

for i, metric in enumerate(metrics):
    plt.subplot(n_rows, 2, i+1)
    sns.barplot(x=kmeans_profiles.index, y=kmeans_profiles[metric])
    plt.title(f'Average {metric.replace("_", " ").title()} by Cluster', fontsize=12)
    plt.xlabel('Cluster')
    plt.ylabel(metric.replace('_', ' ').title())

plt.tight_layout()
plt.show()

# 11. Feature Importance Analysis
print("\n--- Feature Importance Analysis ---")

# Feature importance based on cluster centroids
feature_importance = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=clustering_features
)

# Compute the variance of each feature across clusters
feature_variance = feature_importance.var()
feature_variance = feature_variance / feature_variance.max()  # Normalize

# Sort features by variance (importance)
sorted_features = feature_variance.sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(12, 8))
sorted_features.plot(kind='bar')
plt.title('Feature Importance in Cluster Formation', fontsize=14)
plt.xlabel('Feature', fontsize=12)
plt.ylabel('Relative Importance', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 12. Recommendations and Next Steps
print("\n--- Healthcare Management Recommendations ---")

print("""
Based on the integrated patient segmentation analysis, here are recommended next steps:

1. Targeted Intervention Programs:
   - Develop specific intervention programs for each patient segment identified through K-means and RFM analysis
   - Focus on high-risk and high-cost patient groups for immediate attention

2. Resource Allocation:
   - Allocate healthcare resources according to segment needs
   - Develop preventive care programs for at-risk segments

3. Cost Management Strategies:
   - Implement cost containment strategies for high-cost segments
   - Develop value-based care models for chronic condition patients

4. Patient Engagement:
   - Create customized engagement strategies based on RFM segments
   - Develop retention programs for patients with low recency scores

5. Implementation Plan:
   - Share findings with clinical teams
   - Develop segment-specific care protocols
   - Set up monitoring systems to track intervention effectiveness
""")

# 13. Save Results
# Uncomment to save results
# merged_results.to_csv('healthcare_segmentation_results.csv', index=False)
# kmeans_profiles.to_csv('cluster_profiles.csv', index=False)

print("\nHealthcare Segmentation Analysis Complete!")
```
